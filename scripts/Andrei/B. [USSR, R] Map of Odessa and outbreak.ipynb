```r
df$Exposure_source <- factor(df$Exposure_source, levels = as.character(df$Exposure_source) %>% unique)
df$weight = 1

dfc <- df %>%
  group_by(Exposure_source) %>%
  summarise(volume = sum(weight), pigs = sum(Susceptible)) %>%
  mutate(share_units=volume/sum(volume)*100.0,share_pigs=pigs/sum(pigs)*100.0) %>%
  arrange(desc(volume))
dfc
```

```r
cs = c(3,2.5)
options(repr.plot.width=cs[1],repr.plot.height=cs[2])

ggplot(dfc, aes("", share_units, fill = Exposure_source)) +
    geom_bar(width = 1, size = 1, color = "white", stat = "identity") +
    coord_polar("y") +
    geom_text(aes(label = volume), 
              position = position_stack(vjust = 0.5)) +
    labs(x = NULL, y = NULL, fill = NULL) +
#     guides(fill = guide_legend(reverse = TRUE)) +
    guides(fill = F) +
    theme_classic() +
    theme(axis.line = element_blank(),
          axis.text = element_blank(),
          axis.ticks = element_blank(),
          plot.title = element_text(hjust = 0.5, color = "#666666")) -> p1

ggsave(plot=p1,width=cs[1],height=cs[2],filename="../../figures/draft/AppendixFig-Odessa1977c.pdf",useDingbats=FALSE)

p1
```
